````python
import flywheel

client = flywheel.Client()

proj = client.projects.find_first('label=RBC_PNC')
````

````python
def get_param_from_df(row, param):
    
    subject = proj.subjects.find_first('label="{}"'.format(row['subject']))

    session = subject.sessions.find_first('label="{}"'.format(row['session']))
    
    for acq in session.acquisitions():
        
        files = acq.files
        
        for i,f in enumerate(files):
            
            if f.name == row['filename']:

                acq = client.get(acq.id)
                
                info = acq.files[i]['info']
                
                return info[param]
````

# Extra: ASL & DWI Fieldmaps

We neglected to add the intendedfor to the fieldmaps for DWI and ASL; let's do that now.

````python
def insert_in_json(fname, datadict, dry_run=True):
    
    with open(fname, 'r') as read_file:
        dat = json.load(read_file)

    dat.update(datadict)
    if dry_run:
        print(dat)
    else:
        print('Updating')
        with open(fname, 'w') as read_file:
            json.dump(dat, read_file, indent=4)
````

````python
import glob

json_sidecars = []
niftis = []
root_dir = '/cbica/projects/RBC/RBC_RAWDATA/bidsdatasets/PNC/'

for filename in glob.iglob(root_dir + '**/**', recursive=True):
    
    if filename.endswith('.nii.gz'):
        niftis.append(filename)
    else:
        if 'dataset_description' in filename:
            continue
        elif filename.endswith('.json'):
            json_sidecars.append(filename)
````

Now, for each subject, gather the corresponding dwi files and the fieldmaps

````python
def clean_path(path, subject):
    
    subject_index = path.find(subject)
    
    return path[subject_index+len(subject)+1:]

clean_path('/cbica/projects/RBC/RBC_RAWDATA/bidsdatasets/PNC/sub-997818717/ses-PNC1/dwi/sub-997818717_ses-PNC1_run-02_dwi.nii.gz', 'sub-997818717')
````

````python
import json

def read_json(fpath):
    
    with open(fpath, 'r') as f:
        data = json.load(f)
        
    return data
````

````python
from tqdm import tqdm

for sub in tqdm(dwi_subjects):
    
    print(sub)
    available_dwis = [clean_path(d, sub) for d in dwi if sub in d]
    print(len(available_dwis), 'dwi scans')
    
    current_fmaps = list(set([x for x in fmap_sidecars if sub in x]))
    print(len(current_fmaps), 'fieldmaps')
    
    # get the intendedfor of the fieldmap
    for f in current_fmaps:
        
        json_dat = read_json(f)
        intendeds = json_dat['IntendedFor']
        #print('old intendedfor', intendeds)

        if any([x not in intendeds for x in available_dwis]):
        
            intendeds.extend(available_dwis)
            #print('new intendedfor:', intendeds)
            intendeds = list(set(intendeds))
            json_dat['IntendedFor'] = intendeds
            #print(json_dat)
            
            insert_in_json(f, json_dat, False)
        
    print()
````

````python
for sub in tqdm(asl_subjects):
    
    print(sub)
    available_asl = [clean_path(d, sub) for d in asl if sub in d]
    print(available_asl)
    print(len(available_asl), 'asl scans')
    
    current_fmaps = list(set([x for x in fmap_sidecars if sub in x]))
    print(len(current_fmaps), 'fieldmaps')
    
    # get the intendedfor of the fieldmap
    for f in current_fmaps:
        
        json_dat = read_json(f)
        intendeds = json_dat['IntendedFor']
        #print('old intendedfor', intendeds)

        if any([x not in intendeds for x in available_dwis]):
        
            intendeds.extend(available_asl)
            #print('new intendedfor:', intendeds)
            intendeds = list(set(intendeds))
            json_dat['IntendedFor'] = intendeds
            #print(json_dat)
            
            insert_in_json(f, json_dat, False)
        
    print()
````
